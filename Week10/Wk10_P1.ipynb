```python
from zipfile import ZipFile
import re
import nltk
import docx
import string
from nltk.corpus import stopwords
```

```python
with ZipFile('text_files.zip', 'r') as zipObj:
    zipObj.extractall()
list_of_files = zipObj.namelist()
list_of_files
```

```python
#punctuations that are to be removed
punctuation=list(string.punctuation)
print(punctuation)
```

```python
#stopwords that will be removed
stop_words = set(stopwords.words('english')) 
stop_words
```

```python
#Regular Expression for phone number
regex_phone_no = r"\(?\d{3}\)?[-.\s]\d{3}[-.\s]\d{4}"
```

```python
#Regular Expression for URL
regex_url = r"^https?:\/\/.*[\r\n]*"
```

```python
DataFile = open('outputfile.dat',"a+", encoding = 'utf8')
for each_file in range(len(list_of_files)):
    if list_of_files[each_file].endswith('.txt'):
        with open(list_of_files[each_file],'r',encoding = 'utf8') as file:
            words = file.read().lower()
        print("Current working filename is:",list_of_files[each_file],end="\n")
            
    elif list_of_files[each_file].endswith('.docx'):
        file = docx.Document(list_of_files[each_file])
        result = [each_file.text for each_file in file.paragraphs]
        words = " ".join(result).lower()
        print("Current working filename is:",list_of_files[each_file],end="\n")
           
    else:
        continue
            
    words = sorted(words.split(), key=str.lower)
    while "" in words:
        words.remove("")  
    words = [re.sub('[0-9]', '' , i) for i in words]
#removing stopwords
    words = [w for w in words if not w in stop_words]
#removing URL
    urlremoved = [re.sub(regex_url, '', s, flags=re.MULTILINE) for s in words]
#removing Phone number
    without_phone_no = [re.sub(regex_phone_no, '', s) for s in urlremoved]
# Removing Unique Words
    uniqueWords = []              
    unique_words = set(without_phone_no)
    for word in unique_words:
        uniqueWords.append(word)
#removing punctuation            
    without_punctuation = [''.join(c for c in s if c not in punctuation) for s in uniqueWords]
    without_punctuation = [s for s in without_punctuation if len(s) > 1]
    print(without_punctuation)
    DataFile.write('"The File Name is ' + list_of_files[each_file] + '"' + ": ")
    for words_in_file in without_punctuation: 
        if words_in_file != without_punctuation[-1] :
            DataFile.write(words_in_file + ", ")
        else :                
            DataFile.write(words_in_file + ".")
    DataFile.write("\n")
DataFile.close()
```
