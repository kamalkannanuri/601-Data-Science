## Testing

```python
import pandas as pd
from scipy.stats import ttest_ind, norm
```

```python
Test1 =  norm.rvs(loc=4,scale=10,size=300)
Test1
```

```python
Test2 =  norm.rvs(loc=4,scale=10,size=800)
ttest_ind(Test1, Test2)
#"The results were not statistically significant." 
```

##### Our p-value>0.05 indicates weak evidence against the null hypothesis, so we fail to reject the null hypothesis.

##### Null Hypothesis: The Yearly Statistics in the LSB and CSB have data dependency.
